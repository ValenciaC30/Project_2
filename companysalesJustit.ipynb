```python
#Step 3: Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
#1. Use Seaborn and Matplotlib to create a line plot visualizing sales data over time.
plt.figure(figsize=(12,5))
v1=sns.lineplot(sales,x='month_number',y='total_units',label='Products',marker='o',color='green',linestyle='--')

plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.title('Sales over the Year')
plt.legend(loc='upper left')
```

```python
#2. Visualize the distribution of sales across different regions:
plt.figure(figsize=(12,5))
month=sales['month_number'].tolist()
facecream=sales['facecream'].tolist()
plt.plot(month,facecream,marker='o',label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales across the year-Facecream')
plt.legend(loc='upper left')
plt.show()
```

```python
#3. Visualize the distribution of sales across different regions:
plt.figure(figsize=(12,5))
month=sales['month_number'].tolist()
toothpaste=sales['toothpaste'].tolist()
plt.plot(month,toothpaste,marker='o',color='blue',label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales across the year-Toothpaste')
plt.legend(loc='upper left')
plt.show()
```

```python
#4. Visualize the distribution of sales across different regions:
plt.figure(figsize=(12,5))
month=sales['month_number'].tolist()
facewash=sales['facewash'].tolist()
plt.plot(month,facewash,marker='o',color='red',label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales across the year-Facewash')
plt.legend(loc='upper left')
plt.show()
```

```python
#4. Visualize the distribution of sales across different regions:
plt.figure(figsize=(10,5))
month=sales['month_number'].tolist()
facewash=sales['facewash'].tolist()
facecream=sales['facecream'].tolist()
shampoo=sales['shampoo'].tolist()
bathingsoap=sales['bathingsoap'].tolist()
moisturizer=sales['moisturizer'].tolist()
toothpaste=sales['toothpaste'].tolist()

plt.plot(month,facewash,marker='o',color='red',label='facecream')
plt.plot(month,facecream,marker='o',color='blue',label='facecream')
plt.plot(month,shampoo,marker='o',color='green',label='shampoo')
plt.plot(month,bathingsoap,marker='o',color='orange',label='bathingsoap')
plt.plot(month,moisturizer,marker='o',color='purple',label='moisturizer')
plt.plot(month,toothpaste,marker='o',color='brown',label='toothpaste')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales across the year')
plt.legend(loc='upper left')
plt.show()
```

```python
#6. Create a scatter plot to visualize the relationship between sales and profit.
plt.figure(figsize=(12,5))
v2=sns.scatterplot(sales,x='total_units',y='total_profit',color='red')
plt.title('Relationship between Sales and Profit')
plt.xlabel('Total Units Sold')
plt.ylabel('Total Profit earned')
```

```python
#5.5. Create a bar plot to visualize the sales distribution across different products.
plt.figure(figsize=(15,5))
facewash=sales['facewash'].tolist()
facecream=sales['facecream'].tolist()
shampoo=sales['shampoo'].tolist()
bathingsoap=sales['bathingsoap'].tolist()
moisturizer=sales['moisturizer'].tolist()
toothpaste=sales['toothpaste'].tolist()

month=sales['month_number']

plt.bar(month-0.25,facewash,align='edge',width=0.10,label='facewash')
plt.bar(month-0.15,facecream,align='edge',width=.10,label='facecream')
plt.bar(month-0.10,shampoo,align='edge',width=.10,label='shampoo')
plt.bar(month+0.10,bathingsoap,align='edge',width=.10,label='bathingsoap')
plt.bar(month+0.15,toothpaste,align='edge',width=.10,label='toothpaste')
plt.bar(month+0.25,moisturizer,align='edge',width=0.10,label='moisturizer')
plt.xlabel('sales')
plt.ylabel('month')
plt.title('Sales across different products')
plt.legend(loc='upper left')
plt.show()
```
